```python
def LDR(img2):
    y,u,v = cv2.split(img2)

    y1 = cv2.equalizeHist(y)

    img3 = cv2.cvtColor(y1, cv2.COLOR_GRAY2BGR)

    img4 = cv2.cvtColor(img3, cv2.COLOR_BGR2YUV)

    y_, u_, v_ = cv2.split(img4)

    img5 = cv2.merge([y_,u,v])

    img6 = cv2.cvtColor(img8, cv2.COLOR_YUV2BGR)
    
    img6 = image_resize(img6)
    
    return img6
```

```python
def image_resize(image):
    maxpixel = max(image.shape)
    
    dim = (maxpixel,maxpixel)
    resized = cv2.resize(image, dim, interpolation = cv2.INTER_AREA)
    
    return resized

    
```
